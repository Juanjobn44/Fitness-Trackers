2.- Hay datos que nos interesa analizar basándonos en agrupaciones, para darle un 
sentido a nuestro análisis en base a esa agrupación. Basándonos en las siguientes 
agrupaciones:
• Por tipo de dispositivo
• Por precio de venta. Estableceremos cuatro grupos en base a la media del precio 
de venta de cada tipo de dispositivo:
o Smartwatches con un precio inferior o igual a la media de precios de 
venta de estos dispositivos
o Smartwatches con un precio superior a la media de precios de venta de 
estos dispositivos
o Fitnessbands con un precio inferior o igual a la media de precios de venta 
de estos dispositivos
o Fitnessbands con un precio superior a la media de precios de venta de 
estos dispositivos
• Por marca


```python
#Agrupación por Device Type para agregar comparativa
df_fb = df[df['Device Type']== 'FitnessBand']
df_sw = df[df['Device Type']== 'Smartwatch']
```

```python
#Media Fitnessband

media_fb = df_fb['Selling Price'].mean()

#Agrego nueva columna 'Comparativa' para FitnessBand
condiciones_fb = [
    (df_fb['Selling Price'] < media_fb),
    (df_fb['Selling Price'] >= media_fb)
    ]
valor = ['barato', 'caro']
df_fb['Comparativa'] = np.select(condiciones_fb, valor)
df_fb
```

```python
#Media Smartwatch
media_sw = df_sw['Selling Price'].mean()

#Agrego nueva columna 'Comparativa' para FitnessBand
condiciones_sw = [
    (df_sw['Selling Price'] < media_sw),
    (df_sw['Selling Price'] >= media_sw)
    ]
valor = ['barato', 'caro']
df_sw['Comparativa'] = np.select(condiciones_sw, valor)
df_sw
```

```python
#Junto los dos dataframes
df_junto = df_fb.append(df_sw)
df_junto
```

```python
#Agrupación por Device Type
df_ordenado = df_junto.sort_values('Device Type')
df_tipo = df_ordenado.set_index(['Device Type', 'Comparativa','Brand Name'])
df_tipo
```

```python
#Agrupación por marca
df_ordenado = df_junto.sort_values('Brand Name')
df_marca = df_ordenado.set_index(['Device Type', 'Comparativa','Brand Name'])
df_marca
```

```python
#Estadisticas para fitnesband
df_fb['Comparativa'].value_counts()
```

```python
#Estadisticas para smartwatch
df_sw['Comparativa'].value_counts()
```

```python
#Duración media
df_junto['Ratio Bateria'] = (df_junto['Average Battery Life (in days)'] * 1000) / (df_junto['Selling Price'])
media_bat = df_junto['Ratio Bateria'].mean()


condiciones = [
    (df_junto['Ratio Bateria'] < media_bat),
    (df_junto['Ratio Bateria'] >= media_bat)
    ]
valor = ['inferior','superior']
df_junto['Comparacion Ratio Bateria'] = np.select(condiciones, valor)
df_junto
```

```python
#Seleccionamos los que tienen ratio superior
df_ratio_sup = df_junto[df_junto['Comparacion Ratio Bateria'] == 'superior']
df_ratio_sup
```

```python
#Ordenado por ratio bateria
df_ord = df_junto.sort_values('Ratio Bateria', ascending=False)
```
